5.

(a)  h(x) = \ln(e^{3x-6} - 2) 

· i. Domain:  e^{3x-6} - 2 > 0 \Rightarrow e^{3x-6} > 2 
   3x - 6 > \ln 2 \Rightarrow x > 2 + \frac{\ln 2}{3} 
  Range: all reals  \mathbb{R}  (log can output any real)
· ii. Inverse:  y = \ln(e^{3x-6} - 2) 
   e^y = e^{3x-6} - 2 
   e^{3x-6} = e^y + 2 
   3x - 6 = \ln(e^y + 2) 
   x = \frac{\ln(e^y + 2) + 6}{3} 
  So  h^{-1}(x) = \frac{\ln(e^x + 2) + 6}{3} 
· iii.  h^{-1}(h(x)) = \frac{\ln(e^{\ln(e^{3x-6} - 2)} + 2) + 6}{3} 
   = \frac{\ln(e^{3x-6} - 2 + 2) + 6}{3} = \frac{\ln(e^{3x-6}) + 6}{3} 
   = \frac{3x - 6 + 6}{3} = x 

(b) Limits:

· i.  \lim_{x \to 2} \frac{x^2 - 7x + 10}{x - 2} 
  Factor numerator:  (x-2)(x-5) , cancel  x-2 , get  2 - 5 = -3 
· ii.  \lim_{x \to 0} \frac{x}{e^{\sqrt{x}}} 
  As  x \to 0^+ ,  \sqrt{x} \to 0 ,  e^{\sqrt{x}} \to 1 , so limit =  0/1 = 0 

(c)  \varepsilon-\delta  proof for  \lim_{x \to 4} x^2 = 16 :
Given  \varepsilon > 0 , choose  \delta = \min(1, \varepsilon/9) .
If  |x-4| < \delta , then  |x+4| \le 9  (since  x \in (3,5) ),
 |x^2 - 16| = |x-4||x+4| < (\varepsilon/9) \cdot 9 = \varepsilon .

(b)  f(x) = -1 - \frac{x^2}{2} + \frac{x^4}{4} 
 f'(x) = -x + x^3 = x(x^2 - 1) = x(x-1)(x+1) 
Critical points:  x = 0, \pm 1 
 f(0) = -1 ,  f(1) = -1 - 1/2 + 1/4 = -1.25 ,  f(-1) = -1.25 

· i. Turning points:  (0,-1), (1,-1.25), (-1,-1.25) 
· ii. Intervals:
   f'(x) > 0  on  (-1,0) \cup (1,\infty)  → increasing
   f'(x) < 0  on  (-\infty,-1) \cup (0,1)  → decreasing
· iii. Inflection:  f''(x) = -1 + 3x^2 
  Set  f''(x) = 0 :  3x^2 = 1 \Rightarrow x = \pm \frac{1}{\sqrt{3}} 
  Concave up:  f''(x) > 0  for  x < -1/\sqrt{3}  or  x > 1/\sqrt{3} 
  Concave down:  -1/\sqrt{3} < x < 1/\sqrt{3} 

(c) Fencing: 2400 ft, rectangle borders straight road, no fence along road.
Let length parallel to road =  L , two widths =  W .
Fencing:  L + 2W = 2400 
Area  A = LW = (2400 - 2W)W = 2400W - 2W^2 
Derivative:  A' = 2400 - 4W = 0 \Rightarrow W = 600 
Then  L = 2400 - 1200 = 1200 
Dimensions:  1200 \times 600  ft

(d) Circular track radius 100 m, runner speed 5 m/s, coach 200 m from center.
Distance between them changing when 100 m apart?
Set center O, coach at C (OC = 200), runner at R (OR = 100), angle  \theta  at O.
By law of cosines:  d^2 = 100^2 + 200^2 - 2\cdot 100\cdot 200 \cos\theta 
Given  d = 100 :
 10000 = 10000 + 40000 - 40000\cos\theta 
 0 = 40000 - 40000\cos\theta \Rightarrow \cos\theta = 1 
So they are collinear O-R-C with R between? That’s not possible since OC=200, OR=100, RC=100 means R is on OC, so angle = 0.

Differentiate:  2d \dot{d} = 40000 \sin\theta \, \dot{\theta} 
At  \theta = 0 ,  \sin\theta = 0 , so  \dot{d} = 0 .
